```python
def get_dummies_multi_var(dataframe, column_name, column_values_list):
    get_dummies_encoded = pd.get_dummies(dataframe[column_name], prefix=column_name)
    for i in range(len(column_values_list)):
        get_dummies_encoded[(column_name + "_" + str(column_values_list[i]))] = get_dummies_encoded[(column_name + "_" + str(column_values_list[i]))].astype(int)
    dataframes = [dataframe, get_dummies_encoded]
    dataframe = pd.concat(dataframes, axis=1)
    dataframe = dataframe.drop(column_name, axis=1)
    return dataframe
```
